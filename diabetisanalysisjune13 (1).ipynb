```python
cols_ = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for col in cols_:
    med_val = data[data[col] != 0][col].median()
    data[col] = data[col].replace(0, med_val)
2
```

```python
ip=data.drop(['Outcome'],axis=1)
ip
```

```python
op=data['Outcome']
```

```python
#tain test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ip,op,test_size=0.2)
```

```python
#standard scalar transform
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
alg1=LogisticRegression()
alg1.fit(x_train,y_train)
```

```python
yp=alg1.predict(x_test)
yp
```

```python
df=pd.DataFrame({'y':list(y_test),'prediction':yp})
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm=confusion_matrix(y_test,yp)
cm
```

```python
cm1=ConfusionMatrixDisplay(cm)
cm1.plot()
plt.show()
```

```python
#recall,precesion,f1-score
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
acc=accuracy_score(y_test,yp)
recall=recall_score(y_test,yp)
precision=precision_score(y_test,yp)
f1=f1_score(y_test,yp)
```

```python
print(acc)
print(recall)
print(precision)
print(f1)
```

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
alg2=KNeighborsClassifier(n_neighbors=1)
alg2.fit(x_train,y_train)
```

```python
yp2=alg2.predict(x_test)
```

```python
#recall,precesion,f1-score
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
acc=accuracy_score(y_test,yp2)
recall=recall_score(y_test,yp2)
precision=precision_score(y_test,yp2)
f1=f1_score(y_test,yp2)
```

```python
print(acc)
print(recall)
print(precision)
print(f1)
```

```python
#naive bayes
from sklearn.naive_bayes import GaussianNB
clf=GaussianNB()
clf.fit(x_train,y_train)
yp3=clf.predict(x_test)
```

```python
from sklearn import metrics
acc=metrics.accuracy_score(y_test,yp3)
recall=metrics.recall_score(y_test,yp3)
precision=metrics.precision_score(y_test,yp3)
f1=metrics.f1_score(y_test,yp3)
```

```python
print(acc)
print(recall)
print(precision)
print(f1)
```

```python
#svm
from sklearn import svm
salg=svm.SVC(kernel='rbf',C=1000,gamma=0.01)
salg.fit(x_train,y_train)
yp4=salg.predict(x_test)
```

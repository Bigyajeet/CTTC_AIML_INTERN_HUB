```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
year=np.arange(2001,2005,1)
Domestic=[2258,2441,2491,2352]
Industrial=[2184,2971,3270,3742]
plt.bar(year+0.0,Domestic,width=0.25,color='g')
plt.bar(year+0.25,Industrial,width=0.25,color='r')
plt.xticks([2001,2002,2003,2004],rotation=45)
```

```python
states=['AP','OR','MP']
worker=[100,200,150]
non_workers=[200,100,150]

#x-axis position
x=np.arange(len(states))

#plotting stacked bars

plt.bar(x,worker,color='royalblue',label='workers')
plt.bar(x,non_workers,bottom=worker,color='darkorange',label='non_workers')

#titles and labels
plt.xlabel('states')
plt.ylabel('workers')
plt.title("states vs worker")
plt.xticks(x,states,rotation=45)
plt.legend()
plt.show()
```

```python
#Scatter plot vs bar plot
x2=[3,5,4,7,2,10,7,8]
y2=[2,6,4,9,6,9,3,10]
plt.bar(x2,y2,width=0.05,color='green')

x=[3,5,4,7,2,10,7,8]
y=[2,6,4,9,6,9,3,10]
plt.scatter(x,y,s=300,color='gold',marker='*',edgecolors='orange',linewidths=10)

x1=[3,5,4,7,2,10,7,8]
y1=[2,6,4,9,6,9,3,10]
plt.scatter(x1,y1,color='brown')
plt.title("Scatter plot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()


```

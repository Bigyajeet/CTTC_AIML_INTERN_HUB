```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
np.random.seed(0)  #constant data
points=500
X=np.linspace(-3,3,points) #array
Y=np.sin(X)+np.random.uniform(-0.5,0.5,points)  #graph
```

```python

#deep learning applied
model=Sequential()
model.add(Dense(50,activation='sigmoid',input_dim=1))
model.add(Dense(50,activation='sigmoid'))
model.add(Dense(1))       #output++
```

```python
adam=Adam(learning_rate=0.01)
model.compile(loss='mse',optimizer=adam)
model.fit(X,Y,epochs=50)
```

```python
predictions=model.predict(X)
plt.scatter(X,Y)
plt.plot(X,predictions,'ro')
plt.show()
```

```python
data1=data.drop(["RowNumber",'CustomerId','Surname','Tenure'],axis=1)
data1
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data1.Geography=le.fit_transform(data1.Geography)
le1=LabelEncoder()
data1.Gender=le1.fit_transform(data1.Gender)
```

```python
data1
```

```python
ip=data1.drop(['Exited'],axis=1)
op=data['Exited']
```

```python
ip
```

```python
op
```

```python
from keras.utils import to_categorical
op=to_categorical(op)
op
```

```python
from sklearn.model_selection import train_test_split
xtr,xts,ytr,yts=train_test_split(ip,op,test_size=0.2)
```

```python
xtr.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtr = sc.fit_transform(xtr)
xts = sc.transform(xts)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
model = Sequential()
model.add(Dense(200, input_dim=9, activation='sigmoid'))
model.add(Dense(100, activation='sigmoid'))
model.add(Dense(100, activation='sigmoid'))
model.add(Dense(2, activation='sigmoid'))
model.compile(Adam(learning_rate=0.01),loss='binary_crossentropy',  metrics=['accuracy'])
```

```python
h=model.fit(xtr,ytr,epochs=50)
```

```python
yp=model.predict(xts)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
predictions=np.argmax(yp,axis=1)
true_labels=np.argmax(yts,axis=1)
```

```python
cm=confusion_matrix(true_labels,predictions)
report=classification_report(true_labels,predictions)
print("confusion matrix")
print(cm)
print("classification report")
print(report)
```

```python
from sklearn.metrics import precision_score,recall_score,accuracy_score
precision=precision_score(true_labels,predictions)
recall=recall_score(true_labels,predictions)
accuracy=accuracy_score(true_labels,predictions)
print("Precision:", precision)
print("Recall:", recall)
print("Accuracy:", accuracy)
```

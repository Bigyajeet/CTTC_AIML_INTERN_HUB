```python
#merge

df2=pd.DataFrame({'HpI':[70,80,190,50],
                  'ind_gdp':[750,190,250,889],
                  'index':[2005,2006,2003,2004],})

df3=pd.DataFrame({'HpI':[70,80,90,50],
                  'ind_gdp':[7050,1900,2050,8089],
                  'index':[2005,2006,2003,2004],})
m1=pd.merge(df2,df3,on='HpI')
```

```python
m1
```

```python
#join
df4=pd.DataFrame({
    'A':[1,2,3],'B':[90,87,80]})
df5=pd.DataFrame({'C':[5,6,7],'D':[88,45,50]
})
joindf4=df4.join(df5)
```

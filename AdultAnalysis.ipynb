```python
data1=data.drop(['education_num','hoursperweek','capitalloss','fnlwgt','marital_status'],axis=1)
```

```python
data1
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data1.occupation=le.fit_transform(data1.occupation)
le1=LabelEncoder()
data1.sex=le1.fit_transform(data1.sex)
le2=LabelEncoder()
data1.workclass=le2.fit_transform(data1.workclass)
le3=LabelEncoder()
data1.education=le3.fit_transform(data1.education)
le4=LabelEncoder()
data1.race=le4.fit_transform(data1.race)
le5=LabelEncoder()
data1.native_country=le5.fit_transform(data1.native_country)
le6=LabelEncoder()
data1.relationship=le6.fit_transform(data1.relationship)
le7=LabelEncoder()
data1.salary=le7.fit_transform(data1.salary)

```

```python
data1
```

```python
ip=data1.drop(['salary','age','native_country'],axis=1)
ip
```

```python
op=data1['salary']
```

```python
#tain test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ip,op,test_size=0.3,random_state=42)
```

```python
#standard scalar transform
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
alg1=LogisticRegression()
alg1.fit(x_train,y_train)
```

```python
yp=alg1.predict(x_test)
yp
```

```python
df=pd.DataFrame({'y':list(y_test),'prediction':yp})
```

```python

from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm=confusion_matrix(y_test,yp)
cm
```

```python
cm1=ConfusionMatrixDisplay(cm)
cm1.plot()
plt.show()
```

```python
#svm
from sklearn import svm
salg=svm.SVC(kernel='rbf',C=1000,gamma=0.01,random_state=42)
salg.fit(x_train,y_train)
yp4=salg.predict(x_test)
```

```python
from sklearn import metrics
acc=metrics.accuracy_score(y_test,yp4)
recall=metrics.recall_score(y_test,yp4)
precision=metrics.precision_score(y_test,yp4)
f1=metrics.f1_score(y_test,yp4)
```

```python
print(acc)
print(recall)
print(precision)
print(f1)
```

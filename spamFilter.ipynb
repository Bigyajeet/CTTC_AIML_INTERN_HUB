```python
#impoet packages
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
df=pd.read_csv("spam .csv",encoding='ISO-8859-1')
df
```

```python
df=df.drop(["Unnamed: 2","Unnamed: 3","Unnamed: 4"],axis=1)
```

```python
df
```

```python
df.columns=['labels','data']
```

```python
df['b_labels'] =df['labels'].map({'ham':0,'spam':1})
Y=df['b_labels'].values
df
```

```python
df.groupby('labels').describe()
```

```python
df_train,df_test,Ytrain,Ytest=train_test_split(df['data'],Y,test_size=0.33)
```

```python
df_train.shape
```

```python
tfidf=TfidfVectorizer(decode_error='ignore')
Xtrain=tfidf.fit_transform(df_train)
Xtest=tfidf.transform(df_test)
```

```python
model=MultinomialNB()
model.fit(Xtrain,Ytrain)
print("train score",model.score(Xtrain,Ytrain))
print("test score",model.score(Xtest,Ytest))
```

```python
X=tfidf.transform(df['data'])
df['predictions']=model.predict(X)
df
```

```python
#spam
spam=df[(df['predictions']==1)&(df['b_labels']==1)]['data']
for msg in spam:
             print(msg)
```

```python
#not spam
not_spam = df[(df['predictions'] == 0) & (df['b_labels'] == 0)]['data']
for msg in not_spam:
  print(msg)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

#visualize
def visualize(label):
    words=''
    for msg in df[df['labels']==label]['data']:
        msg=msg.lower()
        words += msg + ''
    wordcloud=WordCloud(width=600,height=400).generate(words)
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
```

```python
visualize('spam')
```

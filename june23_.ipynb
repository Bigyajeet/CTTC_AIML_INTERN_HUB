```python
#capturing image from webcam
cap=cv2.VideoCapture(0)
if cap.isOpened():
    ret,frame=cap.read()
    print(ret)
    print(frame)
else:
    ret=False
image=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.title("My picture")
plt.axis('off')
plt.show()
cap.release()
```

```python
#vdo
vdo=r'v.mp4'
cv2.namedWindow(vdo)
cap=cv2.VideoCapture(vdo)
if cap.isOpened():
    ret,frame=cap.read()
    print(ret)
    print(frame)
else:
    ret=False
while ret:
    ret,frame=cap.read()
    if not ret:
        break
    cv2.imshow(vdo,frame)
    if(cv2.waitKey(30)==27):
        break
cv2.destroyAllWindows()
cap.release()
```

```python
#face detection using haarcsacade
face=r'haarcascade_frontalface_default.xml'
model=cv2.CascadeClassifier(face)
facepoints=model.detectMultiScale(img)
for x,y,w,h in facepoints:
    cv2.rectangle(img,(x,y),(x+w,y+h),(123,45,20),3)
plt.imshow(img[::,::,::-1])
plt.show()
```

```python
face_detec=r'haarcascade_frontalface_default.xml'
cap=cv2.VideoCapture(0)
ret,fram=cap.read()
while ret:
    ret,frame=cap.read()
    print(ret)
    print(frame)
    if not ret:
        break
    model=cv2.CascadeClassifier(face_detec)
    facepoints=model.detectMultiScale(frame)
    
    for x,y,w,h in facepoints:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),3)
        image=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        cv2.imshow('My face',image)
    if cv2.waitKey(10)==27:
        break
cv2.destroyAllWindows()
cap.release()
    
    
```

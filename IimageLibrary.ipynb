```python
img=np.zeros((500,500,3),dtype=np.int32)
```

```python
img.shape
```

```python
plt.imshow(img)
plt.show()
```

```python
import cv2
cv2.rectangle(img,(50,150),(200,50),(255,0,0),3)
plt.imshow(img)
plt.show()
```

```python
#we can create a filled shape by taking thickness=-1
cv2.rectangle(img,(250,150),(400,50),(0,0,255),-1)
plt.imshow(img)
plt.show()
```

```python
cv2.circle(img,(300,300),100,(0,255,0),3)
plt.imshow(img)
plt.show()
```

```python
cv2.line(img,(500,500),(0,0),(255,255,255),5)
plt.imshow(img)
plt.show()
```

```python
#putting txt on image
img=np.zeros((500,500,3),dtype=np.int32)
plt.imshow(img)
plt.show()
```

```python
import cv2
font=cv2.FONT_HERSHEY_PLAIN
cv2.putText(img,'Bigyajeet',(10,400),font,3,(255,255,255),2)
plt.imshow(img)
plt.show()
```

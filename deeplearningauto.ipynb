```python
data2=data.drop(['car_name'],axis=1)
```

```python
#ip/op creation
ip=data2.drop(['mpg'],axis=1)
```

```python
ip
```

```python
op=data2['mpg']
```

```python
op
```

```python
from sklearn.model_selection import train_test_split
xtr,xts,ytr,yts=train_test_split(ip,op,test_size=0.2)
```

```python
ip.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(xtr)
sc.fit(xts)
xtr = sc.transform(xtr)
xts = sc.transform(xts)
```

```python
model=Sequential()
model.add(Dense(50,activation='sigmoid',input_dim=7))
model.add(Dense(30,activation='sigmoid'))
model.add(Dense(1))
```

```python
model.compile(Adam(learning_rate=0.01),loss='mse')
model.fit(xtr,ytr,epochs=500)
```

```python
predictions=model.predict(xts)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
b2=mean_squared_error(yts,predictions)
c2=r2_score(yts,predictions)
print(b2)
print(c2)
```

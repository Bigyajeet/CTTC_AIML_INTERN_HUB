```python
ip=data.drop('quality',axis=1)
```

```python
op=data['quality']
```

```python
from sklearn.model_selection import train_test_split
xtr,xts,ytr,yts=train_test_split(ip,op,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtr=sc.fit_transform(xtr)
xts= sc.transform(xts)
```

```python
#logistic Regression
from sklearn.linear_model import LogisticRegression
alg=LogisticRegression()
alg.fit(xtr,ytr)

yp2=alg.predict(xts)

from sklearn import metrics
cm=metrics.confusion_matrix(yts,yp2)

accuracy=metrics.accuracy_score(yts,yp)
recall=metrics.recall_score(yts,yp,average='macro')
print(accuracy)
print(recall)
```

```python
#naive bayes
from sklearn.naive_bayes import GaussianNB
clf=GaussianNB()
clf.fit(xtr,ytr)

yp3=clf.predict(xts)
```

```python
from sklearn import metrics
recall=metrics.recall_score(yts,yp3,average='macro')
accuracy=metrics.accuracy_score(yts,yp3)
print(recall)
print(accuracy)
```

```python
#svm
from sklearn import svm
salg=svm.SVC(kernel='rbf',C=1000,gamma=0.01)
salg.fit(xtr,ytr)
yp4=salg.predict(xts)
```

```python
from sklearn import metrics
recall=metrics.recall_score(yts,yp4,average='macro')
accuracy=metrics.accuracy_score(yts,yp4)
print(recall)
print(accuracy)
```

```python
#DECISIONTREECLASSIFIER
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(criterion='entropy')

#train decision tree classifier
clf=clf.fit(xtr,ytr)

#predict the response for the dataset
y_pred=clf.predict(xts)

from sklearn import metrics
accuracy2=metrics.accuracy_score(yts,y_pred)
recall2=metrics.recall_score(yts,y_pred,average='macro')
print(recall2)
print(accuracy2)
```

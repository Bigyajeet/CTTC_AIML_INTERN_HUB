```python

colss=['lifetime','broken','pressureInd','moistureInd','temperatureInd','team','provider']
for i, col in enumerate(colss):
    plt.subplot(3, 3, i + 1)
    sns.histplot(data[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()

plt.show()
```

```python

colss=['lifetime','pressureInd','moistureInd','temperatureInd','team','provider']
plt.figure(figsize=(15, 10))
for i, col in enumerate(colss):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(x='broken', y=col, data=data)
    plt.title(f'{col} by broken')
plt.tight_layout()
plt.show()
```

```python
colss=['lifetime','pressureInd','moistureInd','temperatureInd','team','provider']
plt.figure(figsize=(15, 10))
for i, col in enumerate(colss):
    plt.subplot(3, 3, i + 1)
    sns.boxenplot(x='broken', y=col, data=data)
    plt.title(f'{col} by broken')
plt.tight_layout()
plt.show()
```

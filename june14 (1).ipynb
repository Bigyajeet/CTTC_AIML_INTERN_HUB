```python
ip=data.drop(['broken','temperatureInd','moistureInd'],axis=1)
ip
```

```python
op=data['broken']
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([('team',OneHotEncoder(),[2]),
                      ('provider',OneHotEncoder(),[3])],
                     remainder='passthrough')
ip=np.array(ct.fit_transform(ip),dtype=str)
```

```python
#tain test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ip,op,test_size=0.2)
```

```python
#standard scalar transform
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
alg1=LogisticRegression()
alg1.fit(x_train,y_train)
```

```python
yp=alg1.predict(x_test)
yp
```

```python
df=pd.DataFrame({'y':list(y_test),'prediction':yp})
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm=confusion_matrix(y_test,yp)
cm
```

```python
cm1=ConfusionMatrixDisplay(cm)
cm1.plot()
plt.show()
```

```python
#recall,precesion,f1-score
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
acc=accuracy_score(y_test,yp)
recall=recall_score(y_test,yp)
precision=precision_score(y_test,yp)
f1=f1_score(y_test,yp)
```

```python
print(acc)
print(recall)
print(precision)
print(f1)
```

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
alg2=KNeighborsClassifier(n_neighbors=1)
alg2.fit(x_train,y_train)
```

```python
yp2=alg2.predict(x_test)
```

```python
#recall,precesion,f1-score
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
acc=accuracy_score(y_test,yp2)
recall=recall_score(y_test,yp2)
precision=precision_score(y_test,yp2)
f1=f1_score(y_test,yp2)
```

```python
print(acc)
print(recall)
print(precision)
print(f1)
```
